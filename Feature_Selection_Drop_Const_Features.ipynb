# **Feature Selection with Dropping constant Features**


In this step we will be removing the features which have constant features actually not important for solving the problem statement

```python
# For create DataFrame Import pandas
import pandas as pd
```

```python
# Make DataFrame of the given data
data = pd.DataFrame({'A':[1,3,5,1,3,5],
                     'B':[2,4,6,8,0,2],
                     'C':[0,0,0,0,0,0],
                     'D':[1,1,1,1,1,1]})
```

```python
data.head()
```

## **Variance ThresHold**

Feature selector that removes all low-variance features.

This feature selection algorithm looks only at the features (X), not the desired outputs(y), and can thus be used for **unsupervised** learning

```python
# It will zero variance features
from sklearn.feature_selection import VarianceThreshold
thres_var=VarianceThreshold(threshold=0)
thres_var.fit(data)
```

```python
thres_var.get_support()
```

```python
data.columns[thres_var.get_support()]
```

```python
const_columns = [column for column in data.columns
                if column not in data.columns[thres_var.get_support()]]

print(len(const_columns))
```

```python
for feature in const_columns:
  print(feature)
```

```python
data.drop(const_columns, axis=1)
```
